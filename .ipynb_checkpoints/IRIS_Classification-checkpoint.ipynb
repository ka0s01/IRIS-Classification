## Splitting the Data into a Training and Testing Split

### Lets use Stratified Shuffle Split so that the proportions are maintained

```python
X = data.drop(columns ="species")
y = data["species"]
split = StratifiedShuffleSplit(n_splits=1,test_size=0.2)
for trainidx, testidx in split.split(X,y):
    X_train, X_test = X.iloc[trainidx], X.iloc[testidx]
    y_train, y_test = y.iloc[trainidx], y.iloc[testidx]
```

```python
plt.hist(y_train)
plt.show()
```

### Here we can see that the data proportions are maintained in the testing split

## Scaling the Data using Standard Scaler


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Evaluating Performance of KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
clf = KNeighborsClassifier()
clf.fit(X_train_scaled,y_train)
```

```python
knn_acc=clf.score(X_test_scaled,y_test)
```

```python
y_pred_KNN = clf.predict(X_test_scaled)
```

```python
cm_KNN = confusion_matrix(y_test, y_pred_KNN)
disp_KNN = ConfusionMatrixDisplay(confusion_matrix=cm_KNN, display_labels=clf.classes_)
disp.plot()
plt.show()
```

```python
from sklearn.metrics import classification_report
CR_KNN = pd.DataFrame(classification_report(y_test, y_pred_KNN,output_dict=True))
CR_KNN
```

## Analysis
### Confusion Matrix
        - The model correctly classified all Setosa samples.
        - There was 1  misclassification with Versivolor and Virginica - There was some confusion between these classes.
### Classification Report
        - Precision for Setosa and Virginica is 100%
        - Recal for Setosa and Versicolor is is perfect however it is 0.9 for Virginica. Thus the model detected all instances of Setosa and Versicolor but missed some instances of Virginica.
        - The F1 score is high, indicating a good balanec between Precision and Sensitivity.

## Conclusion
    KNN is very good at classifiying Setosa samples which makes sense since as we saw in the scatter plots, the setosa samples have very distinct feature values. The model has some confusion in classifying Virginica and Versicolor, they   often have overlapping features

----

## Evaluation Performance of Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train_scaled,y_train)
```

```python
lr_acc= clf.score(X_test_scaled,y_test)
```

```python
y_pred_LR = clf.predict(X_test_scaled)
```

```python
cm_LR = confusion_matrix(y_test, y_pred_LR)
disp_LR = ConfusionMatrixDisplay(confusion_matrix=cm_LR, display_labels=clf.classes_)
disp.plot()
plt.show()
```

```python
CR_LR = pd.DataFrame(classification_report(y_test, y_pred_LR,output_dict=True))
CR_LR
```

## Analysis
    The confusion matrix is  identical to KNN, except for a marginal drop in precision and recall for versicolor and virginica.
## Conclusion
    Logistic regression achieved 93.3% accuracy, classifying setosa perfectly but confusing one virginica with versicolor, as shown in the confusion matrix.
